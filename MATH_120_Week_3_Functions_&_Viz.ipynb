### Example Pattern
Write small, testable functions with clear docstrings.


```python
# define a funciton
def double():
  return 3*2
```

```python
# define a function that doubles in input
def double_better(x):
  # here is the docstring
  """
  this function doubles the input, x. multiplies the input by 2
  """
  # x is a parameter of the double better fxn
  return x*2
```

```python
double_better('my name ') # doubles the string up. constant * string = more of the same string
double_better(12)
```

```python
help(double_better)
```

```python
def double_best(x):

  """
  this function doubles the input, x. multiplies the input by 2
  """
  # == is to check for equality
  # = is to assign

  # if x is a string, raise a type error
  if type(x) == str:
    raise TypeError("Input cannot be a string")

  if x < 0:
    raise ValueError("Input cannot be negative")

  return x*2
```

```python
double_best(3)
```

```python
double_best(-3)
```

```python
double_best('bus')
```

```python
def affine_transformation(x,a): # multiply by a constant then add a constant
  """
  multiplies by a constant then add a constant
  """
  y = double_best(x) + a
  return y
```

```python
affine_transformation(100,5)
```

```python
# your solution here
def area_of_polygon(n,side_length): # n is the number of sides, side_length is the actual length
  """
  returns the area of a regular n-gon
  """
  return (n * side_length**2)/(4 * (math.tan(math.pi/n)))

def tileable(n,side_length,radius):
  """
  decides whether a regular n-gon of that size could be inscribed in a circle of given radius
  """
  return (side_length/(2*math.sin(math.pi/n))) <= radius
  # A = (n * s²)/(4 * tan(180°/n))
```

## Mini‑Lecture: Plotly Express vs Graph Objects

- **Plotly Express (px)**: High‑level, concise API that maps **data columns** to **visual encodings** (x, y, color, size, facet, etc.). Great for quick, expressive figures.
- **Graph Objects (go)**: Low‑level, fully flexible API where you assemble figures from **traces** and **layouts**. Useful for complex custom visuals and fine control.

**Figure Anatomy**
- `data` (traces): e.g., `go.Scatter`, `go.Bar`, `go.Box`
- `layout`: axes, titles, legends, margins, annotations, templates
- `fig.update_*` methods modify properties after creation


```python
# Graph Objects example: two traces
x = np.linspace(0, 2*np.pi, 200)
y1 = np.sin(x)
y2 = np.cos(x)

fig = go.Figure() # this makes a blank canvas
fig.add_trace(go.Scatter(x=x, y=y1, mode='lines', name='sin(x)')) # add points/lines as traces to canvas
fig.add_trace(go.Scatter(x=x, y=y2, mode='lines', name='cos(x)')) #
fig.update_layout(
    title="Sine and Cosine with Graph Objects",
    xaxis_title="x (radians)",
    yaxis_title="value",
)
fig.show()
```

## Customization Essentials
- Titles and axis labels: `fig.update_layout(title=..., xaxis_title=..., yaxis_title=...)`
- Hover info: `hover_name`, `hover_data`, or `fig.update_traces(hovertemplate="...")`
- Templates (themes): use `template="plotly"`, `"plotly_white"`, `"ggplot2"`, `"seaborn"`, `"plotly_dark"`, etc.
- Legends: `fig.update_layout(legend=dict(...))`
- Color scales: for continuous vars, `color_continuous_scale="Viridis"` (or any named scale).
- Save your figure:
  - Static image (requires `kaleido`): `fig.write_image("figure.png")`
  - HTML: `fig.write_html("figure.html", include_plotlyjs="cdn")`


```python

fig = px.scatter(
    gm1957, x="gdpPercap", y="lifeExp",
    color="continent", size="pop",
    hover_name="country", size_max=50,
    title="Customization Demo"
)
fig.update_layout(
    xaxis_title="GDP per Capita (log)",
    yaxis_title="Life Expectancy (years)",
    template="plotly_white",
    margin=dict(l=40, r=20, t=60, b=40)
)
fig.update_xaxes(type="log")
fig.show()
```


## Bonus: Subplots (Combining Multiple Charts)
When `facet_*` isn't flexible enough, you can combine different chart types with **`make_subplots`** from `plotly.subplots`.


```python

from plotly.subplots import make_subplots

fig = make_subplots(rows=1, cols=2, subplot_titles=("Histogram of Tip", "Boxplot of Tip by Day"))
fig.add_trace(go.Histogram(x=tips["tip"], name="Tip"), row=1, col=1)
fig.add_trace(go.Box(y=tips["tip"], x=tips["day"], name="Tip by Day"), row=1, col=2)
fig.update_layout(title_text="Subplots Example", showlegend=False)
fig.show()
```
